```python
#Karthick 
#24BAD055
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Load dataset
df = pd.read_csv("data.csv", encoding="ISO-8859-1")
```

```python
# Inspect dataset
print(df.head())
print(df.tail())
print(df.info())
print(df.describe())
```

```python
# Check missing values
print("Missing Values:")
print(df.isnull().sum())

# Sales per product
product_sales = df.groupby("Description")["Quantity"].sum().head(10)
```

```python
# Bar chart
plt.figure()
product_sales.plot(kind="bar")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.title("Top Products by Sales")
plt.show()
```

```python
# Line chart
plt.figure()
product_sales.plot(kind="line")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.title("Sales Trend by Product")
plt.show()
```
