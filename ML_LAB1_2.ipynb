```python
# Load dataset
df = pd.read_csv("diabetes.csv")
```

```python
# Explore dataset
print(df.head())
print(df.info())
```

```python
# Check missing values
print("Missing Values:")
print(df.isnull().sum())
```

```python
# Glucose histogram
plt.figure()
plt.hist(df["Glucose"], bins=20)
plt.xlabel("Glucose Level")
plt.ylabel("Frequency")
plt.title("Glucose Distribution")
plt.show()
```

```python
# Age boxplot
plt.figure()
plt.boxplot(df["Age"])
plt.ylabel("Age")
plt.title("Age Distribution")
plt.show()
```
