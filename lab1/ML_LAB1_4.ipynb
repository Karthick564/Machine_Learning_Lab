```python
#Karthick
#24BAD055
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset
df_bank = pd.read_csv("marketing_campaign.csv", sep='\t')
df_bank.head()
```

```python
# Inspect Structure and Missing Valuesdf_bank.info()
df_bank.isnull().sum()
```

```python
#Convert Birth Year to Age  
import datetime
current_year = datetime.datetime.now().year
df_bank['Age'] = current_year - df_bank['Year_Birth']
```

```python
#Visualize Age Distribution
plt.figure()
plt.hist(df_bank["Age"].dropna(), bins=20)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution")
plt.show()
```

```python
#Visualize Income Distribution
plt.figure()
plt.boxplot(df_bank["Income"].dropna())
plt.ylabel("Income")
plt.title("Income Distribution")
plt.show()
```

```python
#Scatter plot
plt.figure()
plt.scatter(df_bank["Income"], df_bank["MntWines"])
plt.xlabel("Income")
plt.ylabel("Amount Spent on Wine")
plt.title("Income vs. Wine Spending")
plt.show()
```
